```python
# 04_pythagorean_theorem

import math  # Import the math library so we can use the sqrt function

def main():
    # Get the lengths of the two perpendicular sides from the user
    ab = float(input("Enter the length of AB: "))  # Length of side AB
    ac = float(input("Enter the length of AC: "))  # Length of side AC

    # Calculate the hypotenuse (BC) using the Pythagorean theorem: BC^2 = AB^2 + AC^2
    bc = math.sqrt(ab**2 + ac**2)  # math.sqrt() computes the square root

    # Output the length of the hypotenuse
    print("The length of BC (the hypotenuse) is: " + str(bc))

if __name__ == '__main__':
    main()
```

```python
# 03_erase_canvas

""" This program implements an 'eraser' on a canvas.

The canvas consists of a grid of blue 'cells' which are drawn as rectangles on the screen.
We then create an eraser rectangle which, when dragged around the canvas, sets all of the rectangles
it is in contact with to white.
"""

from graphics import Canvas
import time

# Define the constants for the canvas and cell sizes
CANVAS_WIDTH : int = 400
CANVAS_HEIGHT : int = 400

CELL_SIZE : int = 40  # Size of each cell on the grid
ERASER_SIZE : int = 20  # Size of the eraser

def erase_objects(canvas, eraser):
    """Erase objects in contact with the eraser."""
    # Get mouse position to know which cells to erase
    mouse_x = canvas.get_mouse_x()
    mouse_y = canvas.get_mouse_y()

    # Calculate the eraser's boundary
    left_x = mouse_x
    top_y = mouse_y
    right_x = left_x + ERASER_SIZE
    bottom_y = top_y + ERASER_SIZE

    # Find all objects that overlap with the eraser
    overlapping_objects = canvas.find_overlapping(left_x, top_y, right_x, bottom_y)

    # Change the color of the overlapping objects (except for the eraser itself) to white
    for overlapping_object in overlapping_objects:
        if overlapping_object != eraser:
            canvas.set_color(overlapping_object, 'white')

# Main function to set up the canvas and the eraser tool
def main():
    # Create the canvas
    canvas = Canvas(CANVAS_WIDTH, CANVAS_HEIGHT)

    # Calculate the number of rows and columns for the grid
    num_rows = CANVAS_HEIGHT // CELL_SIZE
    num_cols = CANVAS_WIDTH // CELL_SIZE

    # Create the grid of blue cells
    for row in range(num_rows):
        for col in range(num_cols):
            left_x = col * CELL_SIZE
            top_y = row * CELL_SIZE
            right_x = left_x + CELL_SIZE
            bottom_y = top_y + CELL_SIZE

            # Create each blue cell
            cell = canvas.create_rectangle(left_x, top_y, right_x, bottom_y, 'blue')

    # Wait for the user to click to start creating the eraser
    canvas.wait_for_click()

    # Get the starting position of the eraser based on the last click
    last_click_x, last_click_y = canvas.get_last_click()

    # Create the eraser as a pink rectangle
    eraser = canvas.create_rectangle(
        last_click_x,
        last_click_y,
        last_click_x + ERASER_SIZE,
        last_click_y + ERASER_SIZE,
        'pink'
    )

    # Move the eraser and erase objects that it touches
    while True:
        # Get the current mouse position
        mouse_x = canvas.get_mouse_x()
        mouse_y = canvas.get_mouse_y()

        # Move the eraser to the mouse position
        canvas.moveto(eraser, mouse_x, mouse_y)

        # Erase anything that the eraser touches
        erase_objects(canvas, eraser)

        # Slow down the loop slightly to make it more responsive and prevent CPU overuse
        time.sleep(0.05)

# Run the main function if this script is executed
if __name__ == '__main__':
    main()
```

```python
# 03_leap_year

def main():
    # Get the year to check from the user
    year = int(input('Please input a year: '))

    # Check if the year is a leap year based on the given criteria
    if year % 4 == 0:  # Checking if the year is divisible by 4
        if year % 100 == 0:  # Checking if the year is divisible by 100
            if year % 400 == 0:  # Checking if the year is divisible by 400
                print("That's a leap year!")
            else:  # The year is divisible by 100 but not by 400
                print("That's not a leap year.")
        else:  # The year is divisible by 4 but not by 100
            print("That's a leap year!")
    else:  # The year is not divisible by 4
        print("That's not a leap year.")


# This line ensures that the main function is called when running the program
if __name__ == '__main__':
    main()
```

```python
# 00_count_nums

def get_user_numbers():
    """
    Create an empty list.
    Ask the user to input numbers and store them in a list.
    Once they enter a blank line, break out of the loop and return the list.
    """
    user_numbers = []
    while True:
        user_input = input("Enter a number: ")

        # If the user enters a blank line, break out of the loop and stop asking for input
        if user_input == "":
            break

        # convert the user input to an integer and add it to the list
        num = int(user_input)
        user_numbers.append(num)

    return user_numbers

def count_nums(num_lst):
    """
    Create an empty dictionary.
    Loop over the list of numbers.
    If the number is not in the dictionary, add it as a key with a value of 1.
    If the number is in the dictionary, increment its value by 1.
    """
    num_dict = {}
    for num in num_lst:
        if num not in num_dict:
            num_dict[num] = 1
        else:
            num_dict[num] += 1

    return num_dict

def print_counts(num_dict):
    """
    Loop over the dictionary and print out each key and its value.
    """
    for num in num_dict:
        print(str(num) + " appears " + str(num_dict[num]) + " times.")

def main():
    """
    Ask the user to input numbers and store them in a list. Once they enter a blank line,
    print out the number of times each number appeared in the list.
    """
    user_numbers = get_user_numbers()  # Get the list of numbers from user input
    num_dict = count_nums(user_numbers)  # Get the count of each number
    print_counts(num_dict)  # Print the counts

# Python boilerplate.
if __name__ == '__main__':
    main()
```

```python
# 01_phonebook

def read_phone_numbers():
    """
    Ask the user for names/numbers to store in a phonebook (dictionary).
    Returns the phonebook.
    """
    phonebook = {}  # Create an empty phonebook dictionary

    while True:
        name = input("Name: ")
        if name == "":
            break  # Break the loop if the user enters an empty name
        number = input("Number: ")
        phonebook[name] = number  # Add name and number to the dictionary

    return phonebook


def print_phonebook(phonebook):
    """
    Prints out all the names/numbers in the phonebook.
    """
    for name in phonebook:
        print(str(name) + " -> " + str(phonebook[name]))  # Print each name and their associated number


def lookup_numbers(phonebook):
    """
    Allow the user to lookup phone numbers in the phonebook
    by looking up the number associated with a name.
    """
    while True:
        name = input("Enter name to lookup: ")
        if name == "":
            break  # Exit the loop if the user enters an empty name
        if name not in phonebook:
            print(name + " is not in the phonebook")  # Print if the name is not found
        else:
            print(phonebook[name])  # Print the phone number for the name


def main():
    """
    Main function to run the phonebook program.
    It reads phone numbers, prints the phonebook, and allows name lookup.
    """
    phonebook = read_phone_numbers()  # Read names and numbers into the phonebook
    print_phonebook(phonebook)  # Print the phonebook
    lookup_numbers(phonebook)  # Allow the user to look up phone numbers


# Python boilerplate to run the main function
if __name__ == '__main__':
    main()
```

```python
# 03_powerful_passwords

from hashlib import sha256

def login(email, stored_logins, password_to_check):
    """
    Returns True if the hash of the password we are checking matches the one in stored_logins
    for a specific email. Otherwise, returns False.

    email: the email we are checking the password for
    stored_logins: a dictionary pointing from an email to its hashed password
    password_to_check: a password we want to test alongside the email to login with
    """

    # Check if the stored password hash for the email matches the hash of the input password
    if stored_logins[email] == hash_password(password_to_check):
        return True
    return False

def hash_password(password):
    """
    Takes in a password and returns the SHA256 hashed value for that specific password.

    Inputs:
        password: the password we want

    Outputs:
        the hashed form of the input password
    """
    # Return the SHA256 hash of the password
    return sha256(password.encode()).hexdigest()

def main():
    # stored_logins is a dictionary with emails as keys and hashed passwords as values
    stored_logins = {
        "example@gmail.com": "5e884898da28047151d0e56f8dc6292773603d0d6aabbdd62a11ef721d1542d8",  # hash for 'password'
        "code_in_placer@cip.org": "973607a4ae7b4cf7d96a100b0fb07e8519cc4f70441d41214a9f811577bb06cc",  # hash for 'Karel'
        "student@stanford.edu": "882c6df720fd99f5eebb1581a1cf975625cea8a160283011c0b9512bb56c95fb"  # hash for 'password123'
    }

    # Testing the login function with various inputs
    print(login("example@gmail.com", stored_logins, "word"))  # Should return False
    print(login("example@gmail.com", stored_logins, "password"))  # Should return True

    print(login("code_in_placer@cip.org", stored_logins, "Karel"))  # Should return True
    print(login("code_in_placer@cip.org", stored_logins, "karel"))  # Should return False

    print(login("student@stanford.edu", stored_logins, "password"))  # Should return False
    print(login("student@stanford.edu", stored_logins, "123!456?789"))  # Should return False

if __name__ == '__main__':
    main()
```

```python
# 01_chaotic_counting

import random

DONE_LIKELIHOOD = 0.3  # The probability that done() will return True

def chaotic_counting():
    for i in range(10):
        curr_num = i + 1  # Start counting from 1 to 10
        if done():  # Check if done() returns True
            return  # If done() returns True, exit the function
        print(curr_num)  # Print the current number

def done():
    """ Returns True with a probability of DONE_LIKELIHOOD """
    if random.random() < DONE_LIKELIHOOD:  # Randomly return True based on the likelihood
        return True
    return False

def main():
    print("I'm going to count until 10 or until I feel like stopping, whichever comes first.")
    chaotic_counting()  # Call the counting function
    print("I'm done")  # Print this once chaotic_counting finishes

if __name__ == "__main__":
    main()
```

```python
# 02_count_even

def count_even(lst):
    """
    Returns number of even numbers in list.
    >>> count_even([1, 2, 3, 4])
    2
    >>> count_even([1, 3, 5, 7])
    0
    """
    count = 0  # Stores the count of even numbers in the list
    for num in lst:  # Loop through the numbers in the list
        if num % 2 == 0:  # If the current number in the list is even (divisible by 2)
            count += 1  # Add one to our count!

    print(count)  # Print out how many even numbers we counted above

def get_list_of_ints():
    """
    Reads in integers until the user presses enter and returns the resulting list.
    """
    lst = []  # Make an empty list to store integers
    user_input = input("Enter an integer or press enter to stop: ")  # Get user input for an integer
    while user_input != "":  # While the user doesn't enter nothing...
        lst.append(int(user_input))  # Cast the user input into an integer and add it to our list
        user_input = input("Enter an integer or press enter to stop: ")  # Get the next user input

    return lst

def main():
    lst = get_list_of_ints()  # Get the list of integers from the user
    count_even(lst)  # Count and print the number of even numbers in the list

# The provided line is required at the end of Python file to call the main() function.
if __name__ == '__main__':
    main()
```

```python
# solution

import random

NUM_ROUNDS = 5

def main():
    print("Welcome to the High-Low Game!")
    print('--------------------------------')

    # Milestone 5: keep track of your score
    your_score = 0

    # Milestone 4: Play multiple rounds
    for i in range(NUM_ROUNDS):
        print("Round", i + 1)
        # Milestone 1: Generate the random numbers and print them out
        computer_num = random.randint(1, 100)
        your_num = random.randint(1, 100)
        print("Your number is", your_num)

        # Milestone 2: Get user input for their choice
        choice = input("Do you think your number is higher or lower than the computer's?: ").lower()

        # Extension 1: Make sure the player inputs a valid choice (higher or lower)
        while choice != "higher" and choice != "lower":
            choice = input("Please enter either higher or lower: ").lower()

        # Milestone 3: Map out all the ways to win the round
        higher_and_correct = choice == "higher" and your_num > computer_num
        lower_and_correct = choice == "lower" and your_num < computer_num

        if higher_and_correct or lower_and_correct:
            print("You were right! The computer's number was", computer_num)
            # Milestone 5: keep track of your score
            your_score += 1
        else:
            print("Aww, that's incorrect. The computer's number was", computer_num)

        # Milestone 5: keep track of your score
        print("Your score is now", your_score)
        print()

    # Extension 2: Conditional ending messages based on performance
    print("Your final score is", your_score)

    if your_score == NUM_ROUNDS:
        print("Wow! You played perfectly!")
    elif your_score > NUM_ROUNDS // 2:
        print("Good job, you played really well!")
    else:
        print("Better luck next time!")

if __name__ == "__main__":
    main()
```

```python
# starter_code

import random

NUM_ROUNDS = 5

def main():
    print("Welcome to the High-Low Game!")
    print('--------------------------------')

    # Initialize the score variable
    your_score = 0

    # Play multiple rounds
    for i in range(NUM_ROUNDS):
        print(f"Round {i + 1}")

        # Generate random numbers for both the user and the computer
        computer_num = random.randint(1, 100)
        your_num = random.randint(1, 100)
        print(f"Your number is {your_num}")

        # Get user input for the guess (higher or lower)
        choice = input("Do you think your number is higher or lower than the computer's?: ").lower()

        # Ensure the user inputs either 'higher' or 'lower'
        while choice not in ["higher", "lower"]:
            choice = input("Please enter either 'higher' or 'lower': ").lower()

        # Check if the guess is correct
        if (choice == "higher" and your_num > computer_num) or (choice == "lower" and your_num < computer_num):
            print(f"You were right! The computer's number was {computer_num}")
            your_score += 1
        else:
            print(f"Aww, that's incorrect. The computer's number was {computer_num}")

        # Display current score
        print(f"Your score is now {your_score}")
        print()

    # Display the final score and a message based on performance
    print(f"Thanks for playing! Your final score is {your_score}")

    if your_score == NUM_ROUNDS:
        print("Wow! You played perfectly!")
    elif your_score > NUM_ROUNDS // 2:
        print("Good job, you played really well!")
    else:
        print("Better luck next time!")

if __name__ == "__main__":
    main()
```

```python
# solution

# Constants for gravitational constants of planets
MERCURY_GRAVITY = 0.376
VENUS_GRAVITY = 0.889
MARS_GRAVITY = 0.378
JUPITER_GRAVITY = 2.36
SATURN_GRAVITY = 1.081
URANUS_GRAVITY = 0.815
NEPTUNE_GRAVITY = 1.14
EARTH_GRAVITY = 1.0

# Constant for Mars weight multiplier
MARS_MULTIPLE = 0.378

def mars_weight():
    """Calculate weight on Mars based on Earth weight"""
    # Get the Earthling's weight from the user
    earth_weight_str = input('Enter a weight on Earth: ')

    # Convert the input string to a float so we can perform calculations
    earth_weight = float(earth_weight_str)

    # Calculate the weight on Mars
    mars_weight = earth_weight * MARS_MULTIPLE

    # Round the result to 2 decimal places
    rounded_mars_weight = round(mars_weight, 2)

    # Print the equivalent weight on Mars
    print('The equivalent weight on Mars: ' + str(rounded_mars_weight))

def planetary_weight():
    """Calculate weight on selected planet based on Earth weight"""
    # Get the Earthling's weight on Earth
    earth_weight = float(input("Enter a weight on Earth: "))

    # Ask the user for the name of the planet
    planet = input("Enter a planet: ")

    # Determine the gravitational constant for the selected planet
    if planet == "Mercury":
        gravity_constant = MERCURY_GRAVITY
    elif planet == "Venus":
        gravity_constant = VENUS_GRAVITY
    elif planet == "Mars":
        gravity_constant = MARS_GRAVITY
    elif planet == "Jupiter":
        gravity_constant = JUPITER_GRAVITY
    elif planet == "Saturn":
        gravity_constant = SATURN_GRAVITY
    elif planet == "Uranus":
        gravity_constant = URANUS_GRAVITY
    else:
        gravity_constant = NEPTUNE_GRAVITY

    # Calculate the equivalent weight on the selected planet
    planetary_weight = earth_weight * gravity_constant

    # Round the result to 2 decimal places
    rounded_planetary_weight = round(planetary_weight, 2)

    # Print the equivalent weight on the selected planet
    print("The equivalent weight on " + planet + ": " + str(rounded_planetary_weight))

def main():
    """Main function to run both Mars and Planetary Weight Calculators"""
    print("Choose an option:")
    print("1. Mars Weight Calculator")
    print("2. Planetary Weight Calculator")
    choice = input("Enter your choice (1 or 2): ")

    if choice == '1':
        mars_weight()
    elif choice == '2':
        planetary_weight()
    else:
        print("Invalid choice. Please select either 1 or 2.")

if __name__ == "__main__":
    main()
```

```python
# handout

# Problem #1: List Practice
def list_practice():
    # Create a list called fruit_list with some fruits
    fruit_list = ['apple', 'banana', 'orange', 'grape', 'pineapple']

    # Print the length of the list
    print("Length of the list:", len(fruit_list))

    # Add 'mango' at the end of the list
    fruit_list.append('mango')

    # Print the updated list
    print("Updated fruit list:", fruit_list)

# Problem #2: Index Game
def access_element(my_list, index):
    """Access the element at the given index."""
    if index < 0 or index >= len(my_list):
        return "Index out of range."
    return my_list[index]

def modify_element(my_list, index, new_value):
    """Modify the element at the given index with a new value."""
    if index < 0 or index >= len(my_list):
        return "Index out of range."
    my_list[index] = new_value
    return my_list

def slice_list(my_list, start, end):
    """Return a slice of the list from start to end index."""
    if start < 0 or end > len(my_list):
        return "Index out of range."
    return my_list[start:end]

def index_game():
    # Initialize a list with mixed elements
    my_list = [10, 'apple', 42, 'banana', 3.14]

    while True:
        print("\nChoose an operation:")
        print("1. Access an element")
        print("2. Modify an element")
        print("3. Slice the list")
        print("4. Exit")

        # Get user input
        choice = input("Enter your choice (1/2/3/4): ")

        if choice == '1':
            # Access an element
            index = int(input("Enter the index to access: "))
            result = access_element(my_list, index)
            print("Access result:", result)

        elif choice == '2':
            # Modify an element
            index = int(input("Enter the index to modify: "))
            new_value = input("Enter the new value: ")
            result = modify_element(my_list, index, new_value)
            print("Modified list:", result)

        elif choice == '3':
            # Slice the list
            start = int(input("Enter the start index: "))
            end = int(input("Enter the end index: "))
            result = slice_list(my_list, start, end)
            print("Sliced list:", result)

        elif choice == '4':
            # Exit the game
            print("Exiting the game.")
            break
        else:
            print("Invalid choice. Please enter 1, 2, 3, or 4.")

def main():
    print("List Practice Exercise:")
    list_practice()  # Call the list practice function

    print("\nIndex Game Exercise:")
    index_game()  # Call the index game function

if __name__ == "__main__":
    main()
```

```python
# solution

# Problem #1: List Practice
def list_practice():
    # Create a list called fruit_list with the given fruits
    fruit_lst = ['apple', 'banana', 'orange', 'grape', 'pineapple']

    # Print the length of the list
    lst_length = len(fruit_lst)
    print("Length of the list:", lst_length)

    # Add 'mango' at the end of the list
    fruit_lst.append('mango')

    # Print the updated list
    print("Updated fruit list:")
    for fruit in fruit_lst:
        print(fruit)

# Problem #2: Index Game
def access_element(lst, index):
    try:
        return lst[index]
    except IndexError:
        return "Index out of range."

def modify_element(lst, index, new_value):
    try:
        lst[index] = new_value
        return lst
    except IndexError:
        return "Index out of range."

def slice_list(lst, start, end):
    try:
        return lst[start:end]
    except IndexError:
        return "Invalid indices."

def index_game():
    lst = [1, 2, 3, 4, 5]  # Example list
    print("Current list:", lst)
    print("Choose an operation: access, modify, slice")

    operation = input("Enter operation: ")

    # Handling the operations based on user input
    if operation == "access":
        index = int(input("Enter index to access: "))
        print(access_element(lst, index))
    elif operation == "modify":
        index = int(input("Enter index to modify: "))
        new_value = input("Enter new value: ")
        print(modify_element(lst, index, new_value))
    elif operation == "slice":
        start = int(input("Enter start index: "))
        end = int(input("Enter end index: "))
        print(slice_list(lst, start, end))
    else:
        print("Invalid operation.")

# Main function to execute both problems
def main():
    print("List Practice:")
    list_practice()

    print("\nIndex Game:")
    index_game()

# Call the main function to run the program
if __name__ == "__main__":
    main()
```
